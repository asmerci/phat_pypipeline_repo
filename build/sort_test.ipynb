```python
# What sort.py does in AWS pipeline
# Make directories titled by the proposal_id and targetname (i.e., 16799_M31-B33-F06-UVIS) - if not already made (for future processing)
# Seperate files with different proposalid and targenames [fields?]
# Move files into the directories with the same proposal id and targname 
# Copy the input data to the proc directory in a dir called Orig_files - wpipe does this!
# Spawn jobs to move process akk of the images
```

```python
def sort_fits(unsorted_dir_path):
    unsorted_targ_list =  glob.glob(unsorted_dir_path + '*.fits')
    for input_filepath in unsorted_targ_list:
        hdu = fits.open(input_filepath)
        filter_name = hdu[0].header['FILTER']
        prop_id = str(hdu[0].header['PROPOSID'])
        target_name = hdu[0].header['TARGNAME']
        rootname = hdu[0].header['ROOTNAME']
        filename = hdu[0].header['FILENAME']
        hdu.close()


        # Write directories directories with
        dir_name = prop_id + '_' + target_name
        field_dir = unsorted_dir_path + '/' + dir_name
        print('Field Directory Path: ', field_dir)
        
        cmd = f"cp {input_filepath} {field_dir}" # Unix command(cmd) to copy files to directory

        # Check if current filename exist and if not create a new dir and copies the current file
        if os.path.exists(field_dir) == True:
            print (f"Copy {input_filepath} to {field_dir}")
            os.system(cmd) # Run Unix to copy
            # print(field_dir, 'Already exist... moving on to next image')
            pass
        else:
            print('Writing directory: ', field_dir)
            os.mkdir(field_dir)
            os.system(cmd) # Run Unix to copy
```

```python
unsorted_dir_path = '/Users/mmckay/phd_projects/phat_pipeline_dev/test_data/'
sort_fits(unsorted_dir_path)
```
